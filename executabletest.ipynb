```python
text=[]

def text_extraction(file_path):
    data = open(file_path, 'r')
    content=data.read()
    print(content)
    content.replace(","," ")
    content.replace('"'," ")
    text.append(content)
    
```

```python
def feature_extraction():
    vectorizer=CountVectorizer(stop_words='english', max_features=1000)
    dtm=vectorizer.fit_transform(text)
    df=pd.DataFrame(dtm.toarray(),columns=vectorizer.get_feature_names())
    df.to_csv(r'malwarefeatures.csv')
```

```python
def calc_frequency(payloads_vectors):
    features_total = 0
    for feature in payloads_vectors:
        features_total += feature
    features_total *= 1.0
    payloads_vectors = [(x/features_total) for x in payloads_vectors]
    return np.array(payloads_vectors, dtype=np.float16)
```

```python
narray=df.to_numpy()
narray = np.vstack([calc_frequency(x) for x in narray])
mean = np.mean(narray,axis=0)


cov=np.cov(np.transpose(narray),bias=False)
invcov=np.linalg.pinv(cov)
invcov.shape
```
